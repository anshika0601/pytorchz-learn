```python
from transformers import AutoTokenizer
#load tokenizer tied to specific model
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
#tokenize
inputs = tokenizer(
    "Hello world!",
    padding='max_length',   # pad shorter sequences to max_length
    truncation=True,        # truncate longer sequences to max_length
    max_length=32,          # desired sequence length
    return_tensors="pt"     # return PyTorch tensors
)


print(inputs.keys())
print(inputs['input_ids'])        #List of vocab IDs
print(inputs['attention_mask'])   #Mask indicating padded vs real tokens
```

```python
from transformers import AutoTokenizer

# 1. WordPiece (BERT)
bert_tok = AutoTokenizer.from_pretrained("bert-base-uncased")
bert_tokens = bert_tok("I love playing football", return_tensors="pt")

# 2. BPE (GPT-2)
gpt2_tok = AutoTokenizer.from_pretrained("gpt2")
gpt2_tokens = gpt2_tok("I love playing football", return_tensors="pt")

# 3. SentencePiece (T5)
t5_tok = AutoTokenizer.from_pretrained("t5-small")
t5_tokens = t5_tok("I love playing football", return_tensors="pt")

# Print results
print("🔹 BERT (WordPiece):", bert_tok.convert_ids_to_tokens(bert_tokens["input_ids"][0]))
print("🔹 GPT-2 (BPE):", gpt2_tok.convert_ids_to_tokens(gpt2_tokens["input_ids"][0]))
print("🔹 T5 (SentencePiece):", t5_tok.convert_ids_to_tokens(t5_tokens["input_ids"][0]))
```

```python
#attention output
from transformers import AutoTokenizer, AutoModel
import torch

# Load model + tokenizer (BERT in this case)
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name, output_attentions=True)

# Example sentence
sentence = "The book that I bought last week was amazing."
inputs = tokenizer(sentence, return_tensors="pt")

# Forward pass with attentions
outputs = model(**inputs)
attentions = outputs.attentions

print(attentions[0][0][3,1])
```

```python
#visualize
from bertviz import head_view


tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])

# Visualize one layer
head_view(attentions, tokens)
```
