text generator

```python
from transformers import pipeline

# Load GPT-2 generator
generator = pipeline("text-generation", model="gpt2")

# Prompt
prompt = "Once upon a time,"
result = generator(prompt, max_length=50, num_return_sequences=1)

print(result[0]['generated_text'])
```

chatbot without conversation history using gpt2

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load GPT-2
model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

model.eval()


```

```python
print("Chatbot ready! Type 'exit' to stop.\n")

while True:
    prompt = input("You: ")
    if prompt.lower() == "exit":
        break

    # Encode input
    inputs = tokenizer.encode(prompt, return_tensors="pt")

    # Generate response
    outputs = model.generate(
        inputs,
        max_length=100,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode output
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print("Bot:", response, "\n")
```

BlenderBot for conversation

```python
from transformers import BlenderbotTokenizer, BlenderbotForConditionalGeneration
```

```python
model_name = "facebook/blenderbot-400M-distill"   # small and fast
tokenizer = BlenderbotTokenizer.from_pretrained(model_name)
model = BlenderbotForConditionalGeneration.from_pretrained(model_name)
```

```python
print("BlenderBot ready! Type 'exit' to quit.\n")

conv_history = []

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break
    #Add conversation history
    conv_history.append("User: " + user_input)


    # Encode input
    context = " ".join(conv_history)
    inputs = tokenizer([context], return_tensors="pt")

    # Generate reply
    reply_ids = model.generate(**inputs)
    reply = tokenizer.decode(reply_ids[0], skip_special_tokens=True)

    print("Bot:", reply, "\n")

    conv_history.append("Bot: " + reply)

```
